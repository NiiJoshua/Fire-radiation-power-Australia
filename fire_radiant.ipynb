```python
k1 = k.drop(['Unnamed: 0','scan','track','acq_date','acq_time','version','instrument'], axis=1)
k1.head()
```

```python
k1.drop(['daynight','satellite'], axis=1, inplace=True)
k1.head()
```

```python
# train dataset
x_train = k1.iloc[:25001, :5]
y_train = k1.iloc[:25001, -1:]

x_test = k1.iloc[25001: , :5]
y_test = k1.iloc[25001:, -1:]
```

```python
X_train = x_train.values
Y_train = y_train.values
```

```python
from sklearn.linear_model import Ridge, Lasso, LinearRegression, ElasticNet
from sklearn.ensemble import RandomForestRegressor
```

```python
models = []
models.append(('LR', LinearRegression()))
models.append(('L', Lasso()))
models.append(('R',Ridge()))
models.append(('EL',ElasticNet()))
models.append(('RF',RandomForestRegressor()))
```

```python
models
```

```python
#
from sklearn.model_selection import cross_val_score, KFold
# Evalueate the models
import warnings
warnings.filterwarnings('ignore')
results = []
names = []
scoring = 'neg_mean_absolute_error'

for name, model in models:
    kfold = KFold(n_splits=10, random_state=1, shuffle=True)
    cv_results = cross_val_score(model,X_train, Y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
# Random Forest Seems to have a better performance
# lets standardize the data
# using pipeline for this
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pipelines = []
pipelines.append(('ScalledLR',Pipeline([('Scaler',StandardScaler()),('LR',LinearRegression())])))
pipelines.append(('ScalledLR',Pipeline([('Scaler',StandardScaler()),('LR',Lasso())])))
pipelines.append(('ScalledLR',Pipeline([('Scaler',StandardScaler()),('LR',Ridge())])))
pipelines.append(('ScalledLR',Pipeline([('Scaler',StandardScaler()),('LR',ElasticNet())])))
pipelines.append(('ScalledLR',Pipeline([('Scaler',StandardScaler()),('LR',RandomForestRegressor())])))
```

```python
# Evaluate each model in turn
results = []
names = []
scoring = 'neg_mean_absolute_error'

for name, model in pipelines:
    kfold = KFold(n_splits=10, random_state=1, shuffle=True)
    cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```
